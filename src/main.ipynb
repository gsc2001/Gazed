```python
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
with open('./cost_matrix.pkl', 'rb') as f:
    cost_matrix = pkl.load(f)
with open('./unary_costs.pkl', 'rb') as f:
    unary_matrix = pkl.load(f)
```

```python
for shot in cost_matrix.keys():
    mat = np.array(cost_matrix[shot])
    plt.plot(mat.flatten())
    print(shot, mat.min(), mat.max(), mat.mean())
```
