```python
import numpy as np
import cv2
import yaml
```

```python
video_path = '../data/afraid_6/afraid_6.mp4'
gaze_path = '../data/afraid_6/gaze/afraid_6_gaze.yml'
```

```python
# read yaml file
with open(gaze_path, 'r') as f:
    gaze = yaml.safe_load(f)

# convert the gaze to numpy array
for key in gaze.keys():
    gaze[key] = np.array(gaze[key], dtype=np.float64)
```

```python
len(gaze[0])
```

```python
video_cap = cv2.VideoCapture(video_path)
gaze_colors = [(0, 0, 255), (255, 0, 0), (0, 255, 0), (0, 255, 255), (255, 0, 255)]
current_gazes = [gaze[key][0] for key in gaze.keys()]
current_gaze_idx = [0 for key in gaze.keys()]
t = 0
while True:
    t += 
    for key in gaze.keys():
        if current_gaze_idx[key] < len(gaze[key]) - 1:
            if gaze[key][current_gaze_idx[key] + 1][2] <= t:
                current_gazes[key] = gaze[key][current_gaze_idx[key] + 1]
                current_gaze_idx[key] += 1
            
    ret, frame = video_cap.read()
    for key in gaze.keys():
        print(current_gazes[key])
        frame = cv2.circle(frame, (int(current_gazes[key][0]), int(current_gazes[key][1])), 10, gaze_colors[key] , -1)
    if ret == False:
        break
    # do something with frame
    cv2.imshow('frame', frame)
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

video_cap.release()

# close window
cv2.destroyAllWindows()
cv2.waitKey(1)
```
